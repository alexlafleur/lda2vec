### Preprocess the data by removing #, @..., http...

```python
tweets = np.loadtxt("../data/tweet.csv", delimiter="-$$$-", dtype=str, comments=None, usecols=None)
```

```python
onlytweet = tweets[:,3]
import re
pattern = r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s!()\[\]{};:'".,<>?]))'''
re_pattern = re.compile(pattern)
tweetnolinks = [re.sub(re_pattern, "", tweet) for tweet in onlytweet]
```

```python
data=np.loadtxt("../data/tweets_no_links.csv", delimiter="-$$$-", comments=None, dtype=str, usecols=4)
data
```

```python
def preprocess_data(data):
    import re
    pattern = r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s!()\[\]{};:'".,<>?]))'''
    re_pattern = re.compile(pattern) #move outside of loop
    sentences = [] 
    for tweet in data:
        
        tweet = re.sub(re_pattern,"", tweet)
        tokens = nltk.word_tokenize(tweet)
        tokens = [word for word in tokens if len(word) > 1]
        sentences.append(tokens)
    return sentences
```

```python
import time
start = time.time()
sentences = preprocess_data(decoded_texts[:50])
time.time() - start
```

```python
sentences[:10]
```

```python
start = time.time()
model = Word2Vec(sentences=sentences, size=300, window=5, min_count=1, workers=4)
print(time.time() - start)
```
