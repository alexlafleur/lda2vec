```python
def create_day_min_data(dateobjlist):
    dateobjlist.sort()
    datedict = {}
    hourdict = {}
    mindict = {}
    for dateobj in dateobjlist:
        day = datetime.datetime.strftime(dateobj, "%Y-%m-%d")
        hour = datetime.datetime.strftime(dateobj, "%Y-%m-%d %H")
        minute = datetime.datetime.strftime(dateobj, "%Y-%m-%d %H:%M")
        if (day in datedict):
            datelist = datedict[day]
            datelist.append(dateobj)
            datedict[day] = datelist
        else:
            datedict[day] = [dateobj]
        if (minute in mindict):
            datelistmin = mindict[minute]
            datelistmin.append(dateobj)
            mindict[minute] = datelistmin
        else:
            mindict[minute] = [dateobj]
        if (hour in hourdict):
            datelisthour = hourdict[hour]
            datelisthour.append(dateobj)
            hourdict[hour] = datelisthour
        else:
            hourdict[hour] = [dateobj]
    return datedict,hourdict, mindict
```

```python
def get_day_min_tweet_amount(datedict, hourdict, mindict):
    tweets_per_day_list = []
    for key, value in datedict.items():
        tweets_per_day = len(datedict[key])
        tweets_per_day_list.append(tweets_per_day)
        
    tweets_per_hour_list = []
    for key, value in hourdict.items():
        tweets_per_hour = len(hourdict[key])
        tweets_per_hour_list.append(tweets_per_hour)

    tweets_per_minute_list = []
    for key, value in mindict.items():
        tweets_per_minute = len(mindict[key])
        tweets_per_minute_list.append(tweets_per_minute)
    return tweets_per_day_list, tweets_per_hour_list, tweets_per_minute_list
```

```python
def tweets_per_unit(userid = "771697761695703040"):
    import datetime
    user_df = all_data[all_data.user_id == userid]
    created_at_list = user_df.created_at_x
    dateobjlist = []
    for date in created_at_list:
        date_obj = datetime.datetime.strptime(date, "%Y-%m-%d %H:%M:%S")
        dateobjlist.append(date_obj)
    datedict,hourdict, mindict = create_day_min_data(dateobjlist)
    return get_day_min_tweet_amount(datedict, hourdict, mindict)
     
```
